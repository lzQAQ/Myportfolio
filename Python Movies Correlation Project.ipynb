```python
# First let's import the packages we will use in this project
# You can do this all now or as you need them
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

#Read in data

df = pd.read_csv(r'/Users/zhangli/Desktop/mini python project/movies.csv')
```

```python
# look at first few rows of the file
df.head()
```

```python
# Check if there is any missing data
for col in df.columns:
    missing_data_percentage = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(missing_data_percentage*100)))
```

```python
# print the data tpyes
print(df.dtypes)
```

```python
df = df.sort_values (by = ['gross'], inplace = False, ascending=False)
```

```python
pd.set_option('display.max_rows',None)
#display the entire data set to see the trend
```

```python
#drop duplicates show all distinct values
df['company'].drop_duplicates().sort_values(ascending = False).head()
```

```python
#build scatter plot between budget and gross revenue
plt.scatter(x= df['budget'], y=df['gross'])
plt.title('Budget vs Gross Revenue')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for film')
plt.show()
```

```python
# Use seaborn plot regplot
sns.regplot(x="gross", y="budget", data=df, scatter_kws = {"color": "red"}, line_kws = {"color": "blue"})
```

```python
#look at correlation data create a correlation matrix
correlation_matrix = df.corr()
```

```python
#create heatmap Visualized the coorelation_matrix we defined
sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
#look at correlation with nonnumeric value (company)
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized.head()
```

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
df_numerized.corr()
```

```python
#unstack each factor to better present the data
correlation_mat = df_numerized.corr()
corr_pair = correlation_mat.unstack()
corr_pair.head()
```

```python
sorted_pairs = corr_pair.sort_values()
sorted_pairs.head()
```

```python
# create high_correlation when correlation between factor and gloss in larger that 0.5
high_corr = sorted_pairs[(sorted_pairs) >0.5]
high_corr
```
